```python
sb.countplot(x=df.JobRole)
plt.xlabel("Job Role")
plt.ylabel("Number of Employess")
plt.title("Distribution of Empolyess acoording to Job Role")
plt.xticks(rotation=90)
plt.show
```

```python
sb.countplot(x=df.Department)
plt.xlabel("Department")
plt.ylabel("Number of Employess")
plt.title("Distribution of Empolyess acoording to Department")
plt.xticks(rotation=90)
plt.show
```

# Distribution of employees across departments insights

1)Maximum employees are under Research and Development department.
2)The Human Resources department has lowest number of employees.
3)Sales department is the second largest, and has around 446 employees.

```python
sb.histplot(x=df.Age)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show
```

# Insights from the Age Distribution Histogram:
1)Most employees are in their 30s, indicating a mid-career workforce.
2)The number of employees decreases sharply after 50, meaning many might be leaving or retiring.

```python
sb.histplot(x=df.MonthlyIncome)
plt.xlabel("Monthly Income")
plt.ylabel("Frequency")
plt.show
```

# Insights from the Monthly Income Histogram
1)Most employees earn less than 7500, meaning many are in lower-paying jobs, which may lead to higher attrition due to dissatisfaction with pay.
2)High-salary employees are fewer, meaning those who get good salaries might stay longer.
3)Very few employees earn above 15,000, which may mean fewer promotions and slow salary growth.

```python
cor=df.corr(numeric_only=True)
cor
```

```python
plt.figure(figsize=(21,9))
sb.heatmap(cor,annot=True)
plt.show()
```

# Insights from the Correlation Matrix of HR Attrition Data
1)Job Level and Monthly Income are strongly related, means high job level then high monthly income.
2)Job Level and Total Working Years have a strong positive correlation, means high job level requird experience employees.
3)Monthly Income and Total Working Years have a strong postiove correlation, means high experience then high Monthly Income.
4)Years in Current Role and Years at Company have a strong positive correlation,
***************************************************************************************
1)Distance From Home and Job Satisfaction are weakly related, meaning living far from work does not strongly affect job satisfaction.
2)Years at Company and Performance Rating have low correlation, meaning staying longer in a company does not always mean a better performance rating.
3)Work-Life Balance and Monthly Income show little connection, meaning a higher salary does not guarantee a better work-life balance.

```python
sb.countplot(x = df.Attrition)
plt.xlabel("Attrition Status")
plt.ylabel("Number Of Employees")
plt.title("Distribution of employees across Job Roles")
plt.xticks(rotation = 90)
plt.show()
```

# Insights from COUNT PLOT of HR Attrition Data 
1)Most employees have stayed in the company, while fewer have left.
2)The number of employees who left is much lower compared to those who stayed.
3)Employee retention is high, but some still choose to leave.

```python
sb.boxplot(x = df.Attrition, y = df.JobSatisfaction)
```

# Insights from Job Satisfaction vs. Attrition
1)Low job satisfaction might be one of the reasons for employees leaving.
2)Employees with the lowest job satisfaction are mostly in the attrition group.
3)Employees who left generally had lower job satisfaction, showing dissatisfaction is a key reason for attrition.

```python
sb.boxplot(x = df.Attrition, y = df.MonthlyIncome)
```

# Insights from Monthly Income vs. Attrition
1)Employees who left had lower salaries, showing pay might be a reason for leaving.
2)Those who stayed earned more on average, meaning better pay may help retain employees.
3)The salary range is bigger for those who stayed, with some earning very high amounts.

```python
sb.boxplot(x = df.Attrition, y = df.PerformanceRating)
```

# Insights from Performance Rating vs. Attrition
1)A few employees have a rating of 4, but they are a small percentage.
2)There is no big difference in performance rating between those who left and those who stayed.
